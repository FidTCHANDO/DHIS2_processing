```python
def collapse(mylist):
    if (isinstance(mylist, str)):
        return mylist
    else:
        if (len(mylist)>=2):
            
            collapsed = mylist[0]
            try:
                for arg in mylist[1:]:
                    collapsed += "|" + arg
                return collapsed
                
            except:
                return collapsed
            
        else:
            print("La liste ne contient pas plusieurs valeurs")
```
